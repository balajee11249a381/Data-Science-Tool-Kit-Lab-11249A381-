```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Step 1 Create sample dataset
data = {
    'StudyHours': [1, 2, 3, 4, 5, 6, 7, 8],
    'ExamScore': [35, 40, 50, 55, 65, 70, 78, 85]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Step 2 Ensure numeric columns (important!)
df['StudyHours'] = pd.to_numeric(df['StudyHours'], errors='coerce')
df['ExamScore'] = pd.to_numeric(df['ExamScore'], errors='coerce')

```

```python
# Step 3 Drop rows with missing values (if any)
df = df.dropna(subset=['StudyHours', 'ExamScore'])
```

```python
# 1 Scatter Plot
plt.scatter(df['StudyHours'], df['ExamScore'])
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Score (Scatter Plot)")
plt.show()
```

```python
# 2 Line Plot
plt.plot(df['StudyHours'], df['ExamScore'], marker='o')
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Exam Score Progress (Line Plot)")
plt.show()
```

```python
# 3 Histogram
plt.hist(df['ExamScore'], bins=5)
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.title("Exam Score Distribution (Histogram)")
plt.show()
```

```python
# 4 Bar Chart
plt.bar(df['StudyHours'], df['ExamScore'])
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Score (Bar Chart)")
plt.show()
```

```python
#10 draw the scatter plot for ant two columns
#(age vs marks)
# Step 1 Load the CSV file
df = pd.read_csv("/content/dataset for pandas(11249a381) - Sheet1.csv")

# Step 2 Ensure numeric columns (convert strings to numbers, invalid values become NaN)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')

# Step 3 Drop rows with missing Age or Marks
df_clean = df.dropna(subset=['Age', 'Marks'])

# Step 4 Draw Scatter Plot
plt.scatter(df_clean['Age'], df_clean['Marks'], color='blue', edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Marks")
plt.title("Age vs Marks (Scatter Plot)")
plt.grid(True)   # Optional: shows grid lines
plt.show()
```

```python
#11 draw the Bar chart for marks
import matplotlib.pyplot as plt

# Ensure 'Marks' is numeric and fill missing values
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce').fillna(0)

# Remove duplicate rows
df = df.drop_duplicates()

# Use the correct column name for x-axis
x_column = 'Roll.no'
y_column = 'Marks'

# Draw bar chart
plt.bar(df[x_column], df[y_column])
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.title("Student Marks Bar Chart")
plt.xticks(df[x_column], df['Name'], rotation=45)  # show names instead of roll numbers
plt.tight_layout()  # prevents label cutoff
plt.show()
```
