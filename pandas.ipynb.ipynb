## A) Creating Dataframes

```python
import pandas as pd
#from a dictionary
data={
    'name':['alice','bob','charlie'],
    'age':[25,30,35],
    'city':['new york','san francisco','los angeles']
}
df=pd.DataFrame(data)
print(df)
#viewing data
print(df.head())
print(df.info())
print(df.describe())
```

B) Selection And Filtering

```python
#selection and filtering
ages=df['age']
#Filtering rows
above_25=df[df['age']>25]
#iloc(integer_based) vs loc(label_based)
row_0=df.iloc[0]#first row(Alice's data)
specific_val=df.loc[0,'city']#'new york'
print(above_25)
```

C)Data cleaning(Handling missing values)

```python
#checking for nulls
print(df.isnull().sum())
#dropping missing values
df_cleaned=df.dropna()
#filling missing values
df_filled=df.fillna(0)#fill NaNs with 0
#or fill with mean
df['age']=df['age'].fillna(df['age'].mean())
print(df['age'])
```

# code after importing csv file

```python
import pandas as pd
df=pd.read_csv('/content/dataset for pandas(11249a381) - Sheet1.csv')
print(df)
```

```python
#2 display only name and mamrks
print(df[['Name','Marks']])
```

```python
#3 some students have missing values in the marks column.replace with 0
df['Marks']=df['Marks'].fillna(0)
print(df)
```

```python
#4 from a student dataset,select students whose marks>60
above_60=df[df['Marks']>60]
print(above_60)
```

```python
#5 sort the students based on marks in descending order
sorted_df=df.sort_values(by='Marks',ascending=False)
print(sorted_df)
```

```python
#6 add a new column result that shows "pass" if marks>=50,else"fail"
df['Result']=df['Marks'].apply(lambda x:'pass' if x>=50 else 'fail')
print(df)
```

```python
#7 count the total number of students in the dataset
total_students=len(df)
print("total_students:",total_students)

```

```python
#8 find the average marks of all students
average_marks=df['Marks'].mean()
print("average_marks:",average_marks)
```

```python
#9 remove duplicate student records from the dataset
df=df.drop_duplicates()
print(df)
```
