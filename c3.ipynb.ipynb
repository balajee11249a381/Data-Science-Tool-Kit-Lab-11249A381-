```python
%%bash
cat << 'EOF' > program.c
#include &lt;stdio.h&gt;

// Function to calculate factorial using recursion
int factorial(int n)
{
    if (n == 0)
        return 1;
    else
        return n * factorial(n - 1);
}
int main()
{
    int num = 5;
    printf("Factorial of %d is %d", num, factorial(num));
    return 0;
}

EOF

gcc program.c -o program
./program
```
