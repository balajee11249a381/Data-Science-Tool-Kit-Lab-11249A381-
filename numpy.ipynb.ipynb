# 1.1.numpy package






```python
#1 array creation
import numpy as np
a=np.array([1,2,3,4,5])
b=np.arange(1,10,2)
c=np.linspace(0,10,5)
d=np.zeros((3,3))
e=np.ones((3,3))
f=np.eye(3)
g=np.random.randint(1,100,size=(3,3))
print("Array a:",a)
print("Arrange b:",b)
print("Linspace c:",c)
print("Zeros d:",d)
print("Ones e:",e)
print("Identity matrix f:",f)
print("Random matrix g:",g)
```

```python
#2 array properties
'''Random matrix g: [[98 19 69]
                  At   [25 36 30]
                     [ 2 58 72]]'''
print("\nShape:",g.shape)
print("Size:",g.size)
print("Data type:",g.dtype)
print("Dimension:",g.ndim)
print("Item size:",g.itemsize)
```

```python
#3 reshape & transpose
'''Random matrix g: [[98 19 69]
                     [25 36 30]
                     [ 2 58 72]]'''
reshaped=g.reshape(1,9)
print("\nReshaped Array:",reshaped)
print("Transpose:\n",g.T)
print("Flattened:",g.flatten())
```

```python
#4 Mathematical Opertions
#Array a: [1 2 3 4 5]
print("multiply:",np.multiply(a,2))
print("square root:",np.sqrt(a))
print("power:",np.power(a,2))
print("absolute:",np.abs([-1,-2,3]))

print("Square:",np.square(a))
print("\nAdd:",np.add(a,2))
print("Subtract:",np.subtract(a,2))
print("Divide:",np.divide(a,2))
```

```python
#5 statistical methods
#Array a: [1 2 3 4 5]
print("Mean:",np.mean(a))
print("Median:",np.median(a))
print("Standard Deviation:",np.std(a))
print("Variance:",np.var(a))
print("Min:",np.min(a))
print("Max:",np.max(a))
print("Sum:",np.sum(a))
print("percentile(20)",np.percentile(a,20))
```

```python
#7 indexing and slicing
#Array a: [1 2 3 4 5]
print("\nfirst element:",a[0])
print("slice:",a[1:4])
'''Random matrix g: [[98 19 69]
                     [25 36 30]
                     [ 2 58 72]]'''
print("2D slice:",g[0:2,1:3])
```

```python
#8 boolean & conditional
'''Random matrix g: [[98 19 69]
                     [25 36 30]
                     [ 2 58 72]]'''
print("\nwhere > 50:",np.where(g>50))
print("any > 50:",np.any(g>50))
print("all > 50:",np.all(g>50))
```

```python
#9 sorting & searching
#Array a: [1 2 3 4 5]
print("\n sorted array a:",np.sort(a))
print("index of max a:",np.argmax(a))
print("index of min a:",np.argmin(a))
print("unique elements:",np.unique([1,2,3,4,4]))
```

```python
#10 linear algebra
x=np.array([[1,2],[3,4]])
y=np.array([[5,6],[7,8]])
print("\nDot product:\n",np.dot(x,y))
print("Inverse:\n",np.linalg.inv(x))
```
